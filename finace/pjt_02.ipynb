# A.데이터 전처리 -데이터 읽어오기

```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('NFLX.csv')
```

```python
df[['Date','Open','High','Low','Close']]
```

# B.데이터 전처리 - 2021년 이후의 종가 데이터 출력

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
start_date = pd.to_datetime('2021-01-01')
selected_date = df[df['Date'] > start_date]
```

```python
plt.plot(selected_date['Date'], selected_date['Close'])
plt.xlabel('Date')
plt.ylabel('Close')
plt.title('NFLX Close Price')
plt.show()
```

# C.데이터 분석 - 2021년 이후 최고, 최저 종가 출력하기

```python
print("최고 종가:",max(selected_date['Close']))
print("최저 종가:",min(selected_date['Close']))
```

# D.데이터 분석 - 2021년 이후 월 별 평균 종가 출력하기

```python
selected_date_1 = selected_date.resample('ME', on='Date')['Close'].mean().reset_index()
```

```python
plt.plot(selected_date_1['Date'], selected_date_1['Close'])
plt.title('Monthly Average Close Price')
plt.xlabel('Date')
plt.ylabel('Average Close Price')
plt.show()
```

# E.데이터 시각화 - 2022년 이후 최고, 최저, 종가 시각화하기

```python
start_date_1 = pd.to_datetime('2022-01-01')
selected_date_2 = df[df['Date'] >= start_date_1]
```

```python
plt.plot(selected_date_2['Date'], selected_date_2['High'], label='High')
plt.plot(selected_date_2['Date'], selected_date_2['Low'], label='Low', color='orange')
plt.plot(selected_date_2['Date'], selected_date_2['Close'], label='Close', color='green')
plt.title('High, Low,and Close Prices since January 2022')
plt.xlabel('Date')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Price')
plt.legend()
```
